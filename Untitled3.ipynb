```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
movies = pd.read_csv(r'C:\Users\sriva\Downloads\movies.csv')     # Movie details: title, genres
ratings = pd.read_csv(r'C:\Users\sriva\Downloads\ratings.csv')   # User ratings for movies
tags = pd.read_csv(r'C:\Users\sriva\Downloads\tags.csv')         # Tags assigned by users
```

```python
print("Movies:", movies.shape)
print("Ratings:", ratings.shape)
print("Tags:", tags.shape)

movies.head()
ratings.head()
tags.head()
```

```python
tag_data = tags.groupby('movieId')['tag'].apply(lambda x: " ".join(x)).reset_index()
movie_data = pd.merge(movies, tag_data, on='movieId', how='left')

# Fill missing tags and combine with genres
movie_data['combined_features'] = movie_data['genres'] + " " + movie_data['tag'].fillna("")
```

```python
cv = CountVectorizer(stop_words='english')
count_matrix = cv.fit_transform(movie_data['combined_features'])
```

```python
similarity = cosine_similarity(count_matrix)
```

```python
def recommend_movie(title):
    title = title.lower()
    # Use contains instead of exact match
    matches = movie_data[movie_data['title'].str.lower().str.contains(title)]
    
    if matches.empty:
        print("‚ùå Movie not found. Please try a different keyword or spelling.")
    else:
        index = matches.index[0]  # Use first matching result
        distances = list(enumerate(similarity[index]))
        sorted_distances = sorted(distances, key=lambda x: x[1], reverse=True)[1:6]
        
        print(f"\nüé¨ Movies similar to '{movie_data.iloc[index]['title']}':")
        for i in sorted_distances:
            print(movie_data.iloc[i[0]]['title'])
```

```python
recommend_movie("Toy Story")
```

```python
# Compute average ratings per movie from the ratings dataset
average_ratings = ratings.groupby('movieId')['rating'].mean()

# Updated recommend function with ratings filter
def recommend_movie(title):
    title = title.lower()
    matches = movie_data[movie_data['title'].str.lower().str.contains(title)]
    
    if matches.empty:
        print("‚ùå Movie not found. Try another keyword.")
        return
    
    index = matches.index[0]
    distances = list(enumerate(similarity[index]))
    sorted_distances = sorted(distances, key=lambda x: x[1], reverse=True)[1:20]  # Check top 20
    
    print(f"\nüé¨ High-rated movies similar to '{movie_data.iloc[index]['title']}':\n")
    count = 0
    for i in sorted_distances:
        movie_id = movie_data.iloc[i[0]]['movieId']
        rating = average_ratings.get(movie_id, 0)
        if rating >= 3.8:  # Only show if average rating is high
            print(f"{movie_data.iloc[i[0]]['title']}  ‚≠ê {round(rating, 2)}")
            count += 1
        if count == 5:
            break
```

```python
recommend_movie("Toy Story")
```

```python
# Average ratings per movie
avg_rating = ratings.groupby('movieId')['rating'].mean()

# Merge with movie titles
top_movies = pd.merge(avg_rating, movies, on='movieId')
top_rated = top_movies.sort_values('rating', ascending=False).head(10)

# Plot
plt.figure(figsize=(10,5))
plt.barh(top_rated['title'], top_rated['rating'], color='skyblue')
plt.xlabel('Average Rating')
plt.title('‚≠ê Top 10 Highest Rated Movies')
plt.gca().invert_yaxis()
plt.show()
```

```python
user_movie_matrix = ratings.pivot_table(index='userId', columns='movieId', values='rating')
```

```python
target_user_id = 1  # You can change this!

# Get the ratings of the target user
target_ratings = user_movie_matrix.loc[target_user_id]

# Compute Pearson correlation with other users
similarities = user_movie_matrix.corrwith(target_ratings, axis=1)
similar_users = similarities.sort_values(ascending=False).drop(target_user_id)
```

```python
# Get ratings of top similar user
top_user = similar_users.index[0]
top_user_ratings = user_movie_matrix.loc[top_user]

# Recommend movies that top user rated highly and target user hasn't watched
unseen_movies = target_ratings[target_ratings.isnull()].index
recommended = top_user_ratings[unseen_movies].sort_values(ascending=False).head(5)

print(f"\nüé¨ Recommended for user {target_user_id} based on similar user {top_user}:\n")
for movie_id in recommended.index:
    movie_title = movies[movies['movieId'] == movie_id]['title'].values[0]
    print(f"{movie_title}  ‚≠ê {recommended[movie_id]:.1f}")
```
